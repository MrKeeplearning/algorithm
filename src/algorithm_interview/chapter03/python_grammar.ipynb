## 2. 리스트 컴프리헨션

### 2.1. map

* map은 리스트의 요소를 지정된 함수로 처리해주는 함수이다.

* 원본 리스트를 변경하지 않고 새로운 리스트를 생성한다.

    ```python
    # 사용법
    list(map(함수, 리스트))
    tuple(map(함수, 튜플))
    ```

````python
# 사용 예시
a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int, a))
a
````

* map에 int와 리스트 a를 넣어 모든 리스트의 요소들을 정수형으로 변환하고 list를 사용해서 다시 결과를 리스트로 만든다.

* map이 반환하는 맵 객체는 이터레이터라서 변수 여러 개에 저장하는 unpacking이 가능하다.

### 2.2. filter

* 반복 가능한 객체에서 특정 조건에 맞는 요소만 가져온다.

* filter에 지정한 함수의 반환값이 True일 때만 해당 요소를 가져온다.

* 사용법
    ```python
    filter(함수, 반복가능한 객체)
    ```

````python
# filter 예제
def f(x):
    return x>5 and x<10

a = range(1, 10)
list(filter(f, a))
````

## 2.3. lambda

* 함수를 생성할 때 사용하는 예약어로 def와 동일한 역할을 한다.

* 함수를 한 줄로 간결하게 표현할 때 사용.

* def를 사용할 정도로 복잡하지 않거나, def를 사용할 수 없을 때 사용.

* 사용법 : `lambda 매개변수1, 매개변수2, ... : 매개변수를 이용한 표현식`

````python
# 두 개의 인수를 받아 서로 더한 값을 돌려주는 lambda 함수
add = lambda a, b : a+b
result = add(3,5)
print(result)
````

## 2.4. 리스트 컴프리헨션

* 기존 리스트를 기반으로 새로운 리스트를 만들어내는 구문.

* 람다 표현식에 map이나 filter를 섞어서 사용하는 것에 비해 가독성이 훨씬 높다.
    ```python
    # 람다 표현식에 map이나 filter를 섞어 사용한 것
    list(map(lambda x: x+10, [1,2,3]))
    ```

````python
# 홀수인 경우 2를 곱해 출력하는 리스트 컴프리헨션
[n*2 for n in range(1, 10+1) if n % 2 == 1]
````

* 버전 2.7 이후로는 리스트 외에도 딕셔너리 등에도 리스트 컴프리헨션을 적용할 수 있게 되었다.

* 대체로 표현식은 2개를 넘지 않아야 가독성을 떨어뜨리지 않고 간결하게 작성할 수 있다.

## 3. 제너레이터

* 루프의 반복 동작을 제어할 수 있는 루틴 형태

* 임의의 조건으로 숫자 1억 개를 만들어내어 계산하는 프로그램이 있을 때, 제너레이터를 사용하지 않는다면 메모리에 숫자 1억개를 보관하고 있어야 한다.

* 제너레이터를 사용하면 제너레이터만 생성하고 필요 시에 숫자를 생성할 수 있다. 1억개의 숫자를 처음부터 모두 만들어 둘 필요가 없다는 것이다.

### 3.1. yield 구문

* yield 구문을 통해서 제너레이터를 리턴 가능.

* 기존 함수는 return을 만나면 값을 리턴하고 모든 함수의 동작을 종료

* yield는 generator가 현재 실행 중이던 중간값을 리턴한 뒤 함수는 종료되지 않은 채 함수가 끝나는 시점까지 실행된다.

````python
def get_natural_number():
    n = 0
    while True:
        n += 1
        yield n

get_natural_number()
````

````python
# 다음 값을 생성하기 위해 next()로 추출
g = get_natural_number()
for _ in range(0, 100):
    print(next(g))
````

#### 참고: python에서 underscore(_)의 의미는?

* 위의 경우와 같이 반복문에서 underscore가 사용되면 변수로 사용될 수 있다.

* 즉, `for i in range(0, 100)` 과 같은 의미이다.

* 이외에도 인터프리터에서 사용될 때, 값을 무시할 때, 긴 숫자의 가독성을 좋게 할 때 등 다양한 목적으로 사용될 수 있다.

## 4. range

* 제너레이터의 방식을 활용하는 대표적인 함수

* for문에서 사용할 경우 내부적으로 제너레이터의 next()를 호출하듯 매번 다음 숫자를 생성

* 파이썬 2.x 버전까지는 숫자를 미리 생성해서 리스트로 리턴하는 방식이었다.
    * 이것은 파이썬 3.x 버전의 `list(range(x))` 와 동일한 결과
    * 제너레이터를 리턴하는 방식은 `xrange()`로 따로 존재
    * 버전 3 이후 xrange는 사라지고 지금의 range만 남게 되었다.

````python
# 숫자 100만개를 생성하는 2가지 방법
a = [n for n in range(1000000)]
b = range(1000000)

# len()으로 길이를 비교하면 둘 다 동일한 100만 개가 출력됨. 비교 연산자 또한 True 리턴.
len(a)
````

````python
len(b)
````

````python
len(a) == len(b)
````

````python
# a에는 이미 생성된 값이 담겨 있으나, b는 생성해야 하는 조건만 존재.
b
````

````python
type(b)
````

````python
sys.getsizeof(a)
````

````python
sys.getsizeof(b)
````

또한, 인덱스로 접근 시 해당 인덱스만큼 바로 생성하도록 되어 있어 리스트와 거의 똑같이 사용하는 것이 가능하다.

## 5. enumerate

여러 가지 자료형(list, set, tuple 등)을 인덱스를 포함한 enumerate 객체로 리턴한다.

````python
a = [1,10,3,6,134,5]
a
````

````python
enumerate(a)
````

````python
list(enumerate(a))
````

이처럼 enumerate는 인덱스를 자동으로 부여해주기 때문에 편리하게 사용할 수 있다.

리스트 a의 인덱스와 값을 함께 출력하려면 어떤 방법을 사용할 수 있을까?

````python
# 일반적인 for문
# a[i] 조회 작업과 전체 길이를 조회하여 루프를 처리하는 형태가 깔끔한 방법은 아니다.
for i in range(len(a)):
    print(i, a[i])
````

````python
# 인덱스를 위한 변수를 별도로 관리하기 때문에 깔끔하지 못함
i = 0
for v in a:
    print(i, v)
    i += 1
````

````python
# enumerate을 활용
for i, v in enumerate(a):
    print(i, v)
````

## 6. 나눗셈 연산자 (`//`)

파이썬 2 이하에서는 `/` 를 사용했을 때 몫만 리턴했다. 그러나 파이썬 3이상부터는 소수점까지 표현된 값을 리턴한다.

`//` 연산자는 파이썬 2이하의 나눗셈 연산자와 동일한 역할을 한다. 즉 몫을 구할 수 있게 한다.

````python
print(5/3)
type(5/3)
````

````python
print(5//3)
type(5//3)
````

````python
print(int(5/3))
type(int(5/3))
````

몫과 나머지를 동시에 구하려면 divmod() 함수를 사용하면 된다.

````python
divmod(5, 3)
````
