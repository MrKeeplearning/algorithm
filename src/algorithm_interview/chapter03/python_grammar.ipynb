## 1. 타입 힌트

CPython의 typing.py에는 선언할 수 있는 타입이 명시되어 있고 아래와 같이 타입 선언이 가능하다.

````python
a: str = "1"
b: int = 1
````

타입 힌트를 사용하지 않는 경우 `def fn(a)` 와 같은 형식을 띄게 되는데, 이 경우 파라미터 a에 int형을 넘겨야 할지 string형을 넘겨야 할지 알기 어렵다.

타입 힌트를 사용하지 않을 때의 문제점

* 가독성 저하

* 버그 유발의 주범

`def fn(a: int) -> bool:` 과 같이 작성하게 되면 파라미터 a가 정수형임을 확인할 수 있고, 리턴 값으로 true 또는 false를 받게 된다는 점도 알 수 있게 된다.

강제 규약은 아니기 때문에 여전히 동적으로 할당될 수 있다. 따라서 아래와 같은 실수를 범하지 않도록 주의가 필요하다.

````python
a: str = 1
type(a)
````

## 2. 리스트 컴프리헨션

### 2.1. map

* map은 리스트의 요소를 지정된 함수로 처리해주는 함수이다.

* 원본 리스트를 변경하지 않고 새로운 리스트를 생성한다.

    ```python
    # 사용법
    list(map(함수, 리스트))
    tuple(map(함수, 튜플))
    ```

````python
# 사용 예시
a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int, a))
a
````

* map에 int와 리스트 a를 넣어 모든 리스트의 요소들을 정수형으로 변환하고 list를 사용해서 다시 결과를 리스트로 만든다.

* map이 반환하는 맵 객체는 이터레이터라서 변수 여러 개에 저장하는 unpacking이 가능하다.

### 2.2. filter

* 반복 가능한 객체에서 특정 조건에 맞는 요소만 가져온다.

* filter에 지정한 함수의 반환값이 True일 때만 해당 요소를 가져온다.

* 사용법
    ```python
    filter(함수, 반복가능한 객체)
    ```

````python
# filter 예제
def f(x):
    return x>5 and x<10

a = range(1, 10)
list(filter(f, a))
````

## 2.3. lambda

* 함수를 생성할 때 사용하는 예약어로 def와 동일한 역할을 한다.

* 함수를 한 줄로 간결하게 표현할 때 사용.

* def를 사용할 정도로 복잡하지 않거나, def를 사용할 수 없을 때 사용.

* 사용법 : `lambda 매개변수1, 매개변수2, ... : 매개변수를 이용한 표현식`

````python
# 두 개의 인수를 받아 서로 더한 값을 돌려주는 lambda 함수
add = lambda a, b : a+b
result = add(3,5)
print(result)
````

## 2.4. 리스트 컴프리헨션

* 기존 리스트를 기반으로 새로운 리스트를 만들어내는 구문.

* 람다 표현식에 map이나 filter를 섞어서 사용하는 것에 비해 가독성이 훨씬 높다.
    ```python
    # 람다 표현식에 map이나 filter를 섞어 사용한 것
    list(map(lambda x: x+10, [1,2,3]))
    ```

````python
# 홀수인 경우 2를 곱해 출력하는 리스트 컴프리헨션
[n*2 for n in range(1, 10+1) if n % 2 == 1]
````

* 버전 2.7 이후로는 리스트 외에도 딕셔너리 등에도 리스트 컴프리헨션을 적용할 수 있게 되었다.

* 대체로 표현식은 2개를 넘지 않아야 가독성을 떨어뜨리지 않고 간결하게 작성할 수 있다.

## 3. 제너레이터

* 루프의 반복 동작을 제어할 수 있는 루틴 형태

* 임의의 조건으로 숫자 1억 개를 만들어내어 계산하는 프로그램이 있을 때, 제너레이터를 사용하지 않는다면 메모리에 숫자 1억개를 보관하고 있어야 한다.

* 제너레이터를 사용하면 제너레이터만 생성하고 필요 시에 숫자를 생성할 수 있다. 1억개의 숫자를 처음부터 모두 만들어 둘 필요가 없다는 것이다.

### 3.1. yield 구문

* yield 구문을 통해서 제너레이터를 리턴 가능.

* 기존 함수는 return을 만나면 값을 리턴하고 모든 함수의 동작을 종료

* yield는 generator가 현재 실행 중이던 중간값을 리턴한 뒤 함수는 종료되지 않은 채 함수가 끝나는 시점까지 실행된다.

````python
def get_natural_number():
    n = 0
    while True:
        n += 1
        yield n

get_natural_number()
````

````python
g = get_natural_number()
for _ in range(0, 100):
    print(next(g))
````

#### 참고: python에서 underscore(_)의 의미는?

* 위의 경우와 같이 반복문에서 underscore가 사용되면 변수로 사용될 수 있다.

* 즉, `for i in range(0, 100)` 과 같은 의미이다.

* 이외에도 인터프리터에서 사용될 때, 값을 무시할 때, 긴 숫자의 가독성을 좋게 할 때 등 다양한 목적으로 사용될 수 있다.
