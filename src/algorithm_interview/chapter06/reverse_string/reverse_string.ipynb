# 문자열 뒤집기

문자열을 뒤집는 함수를 작성하라. 입력값은 문자 배열이며, 리턴 없이 리스트 내부를 직접 조작하라.

입력 : `["h", "e", "l", "l", "o"]`

출력 : `["o", "l", "l", "e", "h"]`

```python
# 입력
value = ["h", "e", "l", "l", "o"]
```

```python
# 투 포인터를 이용한 스왑
def reverse_string1(s: list[str]) -> None:
    left, right = 0, len(s) - 1
    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1
```

```python
reverse_string1(value)
value
```

```python
# 파이썬 내장함수를 활용한 파이썬다운 방식
def reverse_string2(s: list[str]) -> None:
    s.reverse()
```

```python
reverse_string2(value)
value
```

만약 입력되는 값이 문자열일 경우 문자열 슬라이싱을 활용하는 것도 가능함.

```python
str_val = ["h", "e", "l", "l", "o"]

def reverse_string3(s):
    s[:] = s[::-1]

reverse_string3(str_val)
str_val
```

### 참고사항
문자열 슬라이싱에서 s\[:]은 문자열의 처음부터 끝까지를 의미한다.
